# Converting the Data

```python
data['Age'].mean()
```

```python
data['Age'] = np.where(data['Age'].isnull(),32,data['Age'])
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
data['Sex'] = label_encoder.fit_transform(data['Sex'])
```

```python
data.drop(["PassengerId","Name","Ticket","Cabin"],axis = 1,inplace = True)
```

```python
data
```

# Finding the K value with the Highest Accuracy

```python
y = data['Pclass']
X = data.drop(['Pclass'], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3,random_state = 0)
X_test
```

```python
k_vals = []
acc_vals = []
for k in range(1, 268):
    k_vals.append(k)
    knn = neighbors.KNeighborsClassifier(n_neighbors=k)
    acc_vals.append(knn.fit(X_train, y_train).score(X_test, y_test))
  
```

```python
k_data = pd.DataFrame(list(zip(k_vals,acc_vals)), columns=['k','Accuracy'])
k_data.head()
```

```python
k_data.sort_values("Accuracy", axis = 0, ascending = False, inplace = True)
```
