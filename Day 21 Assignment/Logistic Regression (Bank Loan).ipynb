```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

# Reading the Data

```python
data = pd.read_excel('Bank_Personal_Loan_Modelling.xlsx' ,sheet_name= 0)
```

# Loading X and Y values

```python
y = data['Personal Loan']
```

```python
x = data[['Age','Experience','Income','CCAvg','Education','Mortgage']]
x
```

```python
x1 = sm.add_constant(x) 
```

```python
Logistic= sm.Logit(y,x1)
```

```python
result=Logistic.fit()
```

# Result

```python
result.summary()
```

Hence we can say that
