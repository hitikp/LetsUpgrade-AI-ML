# Converting the Data

```python
data.Attrition = data.Attrition.replace('No', 0)
data.Attrition = data.Attrition.replace('Yes', 1)
```

```python
data.Over18 = data.Over18.replace('N', 0)
data.Over18 = data.Over18.replace('Y', 1)
```

```python
data.head()
```

# Loading X and Y values

```python
y = data.Attrition
y
```

```python
x = data[['Age','DistanceFromHome','Education','JobLevel','MonthlyIncome','NumCompaniesWorked','PercentSalaryHike','TotalWorkingYears','TrainingTimesLastYear','YearsAtCompany','YearsWithCurrManager']]
```

```python
x
```

```python
x1 = sm.add_constant(x) 
```

```python
Logistic = sm.Logit(y,x1)
```

```python
result=Logistic.fit()
```

# Result
